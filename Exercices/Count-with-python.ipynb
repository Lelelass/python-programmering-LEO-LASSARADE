## 3. Classification accuracy

```python
def accuracy (TP: int, TN: int, FP: int, FN: int):
    return (sumTPTN := TP + TN) / (sumTPTN + FP + FN) # Walrus fungerar endast från python 3.8. Ville prova denna för att begränsa kalkuler.

print(f"Accuracy: {accuracy(2,985,2,11)}")
```

Accuracy här är relativt högt, men när det gäller en applikation så känsligt som att känna igen eld och brand så är kanske modelen inte tillräckligt bra än, man behöver högre accuracy eftersom riskerna är enorma. Med 11 False negative, så hade 11 brander gått oupptäckta.

## 5. Euclidian distance

```python
import math

def euclidian(P1: tuple, P2: tuple):
    return math.sqrt((P2[0]-P1[0])**2 + (P2[1]-P1[1])**2)

print(f"Euclidian distance : {euclidian((3,5),(-2,4)):.2f}")
```

## 6. Euclidian distance 3D

```python
import math

def euclidian_3D(P1: tuple, P2: tuple):
    return math.sqrt((P2[0]-P1[0])**2 + (P2[1]-P1[1])**2 + (P2[2]-P1[2])**2)

print(f"Euclidian distance : {euclidian_3D((2,1,4),(3,1,0)):.2f}")
```
