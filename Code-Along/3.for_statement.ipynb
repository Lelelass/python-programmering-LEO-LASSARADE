# For loop

## Nationella prov - pasta uppgift

```python
import math

for year in range(400):
    pasta_consumption = .791 * math.exp(.0526*year)
    if pasta_consumption >= 15:
        print(f"År {year+1960} är pastakonsumptionen 15 kg per år per person")
        break

```

```python
import matplotlib.pyplot as plt

pasta_per_year = []

for year in range(140):
    pasta_consumption = .791 * math.exp(.0526*year)
    pasta_per_year.append(pasta_consumption)

plt.plot(range(1960,2100), pasta_per_year)
plt.xlabel("År")
plt.ylabel("Pastakonsumption i kg")
plt.title("Pastakonsumtionsmodell per kg per person från 1960")
```
