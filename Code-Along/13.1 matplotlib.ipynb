```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5)
f = np.sin(x)

print(x)
```

MATLAB Style interface

```python
plt.figure(dpi = 100)
plt.plot(x, f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.figure(dpi = 100, figsize =(10,3))
plt.subplot(1,2,1) # r(rows, cols, panel number)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
```

## OOP approach

```python
fig = plt.figure(dpi = 100)
#ax object can have a plot, axis, labels, title,...
ax = plt.axes()
print(fig)

ax.plot(x,f)
ax.set(xlabel = "x", ylabel = "y", title = "$\sin(x)$", xlim =(0,5))

```

```python
fig, ax = plt.subplots(1,2, dpi = 100, figsize =(10,4))
rng = np.random.RandomState(42)
data_normal_dist = rng.randn(1000)
ax(0).hist(data_normal_dist, bins = 10, alpha = 0.5)
```
